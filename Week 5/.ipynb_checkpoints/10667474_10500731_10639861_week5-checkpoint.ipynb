```python
# Hier wordt voor elke taal gekeken of het een macro language is, zo ja dan wordt deze omgezet naar de juist verzameltaal
# Hier enkel voor arabisch gedaan, omdat het handmatig moet en heel veel werk is om het voor elke taal te doen
def macro_change(lang):
    ara = ["aao","abh","abv","acm","acq","acw","acw","acx","acy","adf","aeb","aec","afb","ajp","apc","apd","arb","arq"]
    if lang in ara:
        return "ara"
    else:
        return lang
```

# Compute our own PhiCorrelation and T-statistic

Maak een functie `PhiCorrelation(i,j)` die gegeven 2 talen, de $\phi$ correlatie en de $t$-statistic uitrekent volgens vergelijkingen [1] en [2] uit het artikel.

* Let vooral op hoe je $N$ bepaalt. En beschrijf goed waarom je precies die keuze hebt genomen. Je uitkomsten hangen daar sterk vanaf.

```python
# Compute our own PhiCorrelation and T-statistic
def PhiCorrelation(i,j): 
    #Your code here
    Mij = edges.loc[(edges['SourceLanguageCode'] == i) & (edges['TargetLanguageCode'] == j), 'Coocurrencesmy'].values[0]
    Mi = stats.loc[stats["Code"]==i, 'Multilingualsmy'].values[0]
    Mj = stats.loc[stats["Code"]==j, 'Multilingualsmy'].values[0]
    N = int(stats.Editorsmy.sum(axis=0)) 
    
    high = (Mij*N) - (Mi*Mj)
    low = (Mi*Mj)*(((N-Mi)*(N-Mj)))
    # zet negatieve waardes om naar positief
    if low < 0:
        low = low *-1 
    PhiCorrelation = high/sqrt(low)
    D = max(Mi,Mj)
    Tstatistic = PhiCorrelation*sqrt(D-2) / sqrt(1 - PhiCorrelation**2)
    print (Mij,Mi, Mj, N, PhiCorrelation, Tstatistic)

# test
PhiCorrelation('afr','nld')
```
